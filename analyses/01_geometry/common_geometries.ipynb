```python
tetrahedral = jnp.array([
    [0, 0, 1],
    [0.943, 0, -0.333],
    [-0.471, 0.816, -0.333],
    [-0.471, -0.816, -0.333]
])

trigonal_planar = jnp.array([
    [1, 0, 0],
    [-0.5, jnp.sqrt(3)/2, 0],
    [-0.5, -jnp.sqrt(3)/2, 0]
])

key = random.PRNGKey(0)
tetrahedral_noised = tetrahedral + random.normal(key, tetrahedral.shape) * 0.1
```

```python
spectra.visualize_geometry(tetrahedral_noised)
```

```python
bispectrum = spectra.Spectra()
tetra_bis = bispectrum.compute_geometry_spectra(tetrahedral)
tetra_bis_noised = bispectrum.compute_geometry_spectra(tetrahedral_noised)
trigonal_planar_bis = bispectrum.compute_geometry_spectra(trigonal_planar)
```

```python
dist1 = jnp.linalg.norm(tetra_bis - tetra_bis_noised)
dist2 = jnp.linalg.norm(tetra_bis - trigonal_planar_bis)
```
